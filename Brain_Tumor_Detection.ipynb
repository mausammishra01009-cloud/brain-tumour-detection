```python
import gradio as gr
import tensorflow as tf
from PIL import Image
import numpy as np
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout

# 1. Initialize an Enhanced Transfer Learning Brain Engine
print("🧠 Constructing Enhanced MobileNetV2 Neural Architecture...")
# Load MobileNetV2 as a base feature extractor pre-trained on ImageNet
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(150, 150, 3))
base_model.trainable = False  # Freeze baseline layers to protect pre-trained weights

# Append our custom medical classification head layers
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)  # Prevents training memorization overfitting
outputs = Dense(4, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=outputs)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
print("✅ Success: Enhanced Transfer Learning Engine Synchronized!")

class_labels = ['Glioma', 'Meningioma', 'No Tumor', 'Pituitary']

# 2. Advanced Evaluation Logic Pipeline
def enhanced_predict(input_image):
    if input_image is None:
        return "Please upload a valid brain MRI scan image.", {}

    # Standardize input frame matrices
    img = Image.fromarray(input_image.astype('uint8'), 'RGB')
    img = img.resize((150, 150))
    img_array = np.array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    # Compute clean floating-point prediction values directly
    predictions = model.predict(img_array)[0]

    # Track dominant diagnostic target parameters
    top_class_idx = np.argmax(predictions)
    top_confidence = float(predictions[top_class_idx] * 100)
    top_label = class_labels[top_class_idx]

    # Construct float dictionary mappings (Feeds the visual Gradio Graph)
    result_dict = {class_labels[i]: float(predictions[i]) for i in range(len(class_labels))}

    # Process Safety Threshold Diagnostics Text Output
    if top_confidence < 50.0:
        diagnosis_summary = (
            f"⚠️ UNCERTAIN PREDICTION WARNING\n\n"
            f"• Top AI Suspect: {top_label.upper()}\n"
            f"• Peak Category Score: {top_confidence:.2f}%\n\n"
            f"CRITICAL VALIDATION ALERT: The top classification probability is under our 50% safety benchmark. "
            f"This indicates structural ambiguities within the scan matrix. Do not use for clinical diagnostic reference."
        )
    else:
        diagnosis_summary = (
            f"⚡ HIGH CONFIDENCE DIAGNOSIS\n\n"
            f"• Top AI Classification: {top_label.upper()}\n"
            f"• Peak Category Score: {top_confidence:.2f}%\n\n"
            f"SYSTEM VALIDATION: The classification metric has successfully crossed the 50% security threshold."
        )

    return diagnosis_summary, result_dict

# 3. Premium Dark Neon UI Theme Customization
neon_theme = gr.themes.Default(
    primary_hue="cyan",       # Vivid Neon Cyan highlights (#00E5FF)
    secondary_hue="purple",   # Bold Purple interface boundaries
    neutral_hue="slate"
).set(
    body_background_fill="#0B1020",        # Deep immersive space theme
    body_text_color="#FFFFFF",             # High contrast text readability
    block_background_fill="#151D35",       # Module containers background
    block_title_text_color="#00E5FF",      # Section title neon markers
    block_label_text_color="#00FF88",      # Functional green accent metadata tags
    button_primary_background_fill="#7C3AED", # Ultraviolet processing button
    button_primary_text_color="#FFFFFF"
)

# 4. Construct Advanced Application Blocks Layout
with gr.Blocks(theme=neon_theme, title="NeuroScan AI Pro") as demo:
    gr.Markdown("<h1 style='text-align: center; color: #00E5FF;'>🧠 NeuroScan AI Pro: Advanced Tumor Detection</h1>")
    gr.Markdown("<p style='text-align: center; color: #A0AEC0;'>Decision-Support Framework Powered by Pre-trained MobileNetV2 Networks</p>")

    with gr.Row():
        with gr.Column(scale=1):
            input_img = gr.Image(label="Patient Brain MRI Input File")
            submit_btn = gr.Button("EXECUTE QUANTUM ANALYSIS", variant="primary")

        with gr.Column(scale=1):
            output_text = gr.Textbox(label="Analysis Summary & Threshold Validation Engine", lines=6)
            # Gradio automatically renders this label component as a sleek horizontal bar chart!
            output_label = gr.Label(label="Full Probability Distribution (Interactive Graph)", num_top_classes=4)

    submit_btn.click(fn=enhanced_predict, inputs=input_img, outputs=[output_text, output_label])

    gr.Markdown(
        "<div style='border: 1px solid #7C3AED; padding: 15px; border-radius: 8px; margin-top: 20px; background-color: #1A1230;'>"
        "<p style='color: #FF4D4D; font-weight: bold; margin: 0;'>⚠️ FORMAL MEDICAL DISCLAIMER:</p>"
        "<p style='color: #CBD5E0; font-size: 13px; margin: 5px 0 0 0;'>"
        "This deep learning framework is created strictly for academic research and educational evaluation. It does not provide medical counseling, "
        "clinical diagnostic tracking, or actionable treatment blueprints. All medical imagery requires definitive verification by certified "
        "radiology specialists and board-certified neurological teams before rendering clinical conclusions.</p>"
        "</div>"
    )

# 5. Launch the Final Upgraded Deployment Web Link
demo.launch(share=True)
```
